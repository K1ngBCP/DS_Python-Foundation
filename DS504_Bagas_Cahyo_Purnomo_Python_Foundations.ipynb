# 2.Hitunglah BMI setiap siswa dengan rumus BMI=(Berat badan (kg))/((Tinggi Badan (m))^2 )

```python
Dataku['BMI'] = Dataku['Berat Badan (Kg)'] / ((Dataku['Tinggi Badan (Cm)']/100)**2)
Dataku
```

Ini adalah cara mengghitung BMI tanpa membuat fungsi. Dimana Dataku akan ditambahkan koom bernama BMI dimana isinya adalah dataku bb dibagi dengan dataku tb yang terlebih dulu : 100 (agar jadi m) lalu hasilnya dikuadratkan. Lalu disini saya juga menampilkannya untuk melihat perbedaannya

# 3.	Buatlah program untuk menghitung BMI!

```python
def BMIKU(Berat_Badan,Tinggi_Badan):
    tinggi_m = Tinggi_Badan / 100
    bmi = Berat_Badan/ (tinggi_m ** 2)
    return bmi
```

Lalu disini ada cara menghitung BMI berbeda dengan sebelumnya, disini saya menggunakan function yang saya beri nama BMIKU, dimana fungsinya membutuhkan 2 inputan yakni bb dan tb, lalu tb akan dibagi 100 agar satuannya menjadi m, lalu bmi = bb dibagi tb yang sudah menjadi m (dibagi 100), lalu akan di return atau dikembalikan hasil bmi tadi

# 4.	Update dataframe df dengan menambahkan hasil perhitungan soal nomor 2, beri nama kolom BMI.

```python
Dataku['BMI1'] = Dataku.apply(lambda row: BMIKU(row['Berat Badan (Kg)'], row['Tinggi Badan (Cm)']), axis=1)
Dataku
```

Lalu disini saya membuat kolom baru BMI1 dimana kolom ini didapat dari memanggil fungsi tadi yakni BMIKU, denggan menggunakn bantuan apply, lalu saya mennggunakan lambda row untuk prosesnya, dimana inputan yang dimasukkan adalah Berat Badan (Kg)untuk berat_badan dan Tinggi Badan (Cm) untuktinggi_badan. Saya juga memanggil dataku untuk menampilkan hasil fungsi ini, dan menunjukkan bahwa hasilnya sama dengan kode yangbukan fungsi

```python
Dataku.drop('BMI1', axis=1, inplace=True)
Dataku
```

# 5.	Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.
●	Normal: 18.5-24.9
●	Overweight: 25.0-29.9
●	Obesity: 30.0-34.9
●	Underweight: < 18.5


```python
def statusbmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25.0:
        return 'Normal'
    elif 25.0 <= bmi < 30.0:
        return 'Overweight'
    elif 30.0 <= bmi < 35.0:
        return 'Obesity'
    else:
        return '-'
```

Disini saya membuat fungsi dengan nama statusbmi dimana isinya berfungsi untuk memberikan kategori siswa berdasarkan BMInya, disini saya menggunakan if elif. Dimana jika BMI kurang dari 18,5 maka akan direturn dengan underweight, lalu jika BMInya 18,5 dan dibawah 25,0 maka akan mereturn atau mengembalikan kategori normal, lalu jika BMInya 25,0 dan dibawah 30,0 maka akan menampilkan atau mereturn overweight, Lalu jika BMI nya berada di antara 30,0 sampai dengan 34,9 akan mereturn obesity, lalu untuk angka atau nilai lainnya (otomatis diatas 34,9) maka akan mereturn - (disoal tidakada ketentuan)

```python
Dataku['Status BMI'] = Dataku['BMI'].apply(statusbmi)
Dataku
```

Lalu disini saya membuat kolom baru yang saya beri nama Status BMI yang isinya didapat dari mengapply fungsi statusbmi tadi dengan ipnutan BMI. Lalu disini saya menampilkan Dataku untuk melihat perbedaannya
